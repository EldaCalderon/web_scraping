```python
# Funci√≥n para enviar el correo con los resultados
def send_email(facturas_descargadas, facturas_fallidas):

    
    # Lista de destinatarios
    recipient_list = ["ecalderon@maxwarehouse.com"]
    
    # Crear la lista numerada de facturas descargadas
    facturas_descargadas_str = "\n".join([f"{index}. {factura}" for index, factura in enumerate(facturas_descargadas, 1)])
    
    # Crear la lista numerada de facturas fallidas
    facturas_fallidas_str = "\n".join([f"{index}. {factura}" for index, factura in enumerate(facturas_fallidas, 1)])

    # Contenido del correo
    subject = "Reporte de Descarga de Facturas"
    body = f"""
    Facturas descargadas correctamente:
    {facturas_descargadas_str}

    Facturas con error al descargar:
    {facturas_fallidas_str}
    """
    
    # Crear el mensaje
    msg = MIMEText(body)
    msg["Subject"] = subject
    msg["From"] = sender_email
    msg["To"] = ", ".join(recipient_list)
    
    try:
        # Conectar al servidor SMTP
        server = smtplib.SMTP("smtp.office365.com", 587)
        server.starttls()  # Usar TLS para seguridad
        server.login(sender_email, sender_password)
        server.sendmail(sender_email, recipient_list, msg.as_string())
        server.quit()
        print("‚úÖ Correo enviado correctamente")
    except Exception as e:
        print(f"‚ö†Ô∏è Error al enviar el correo: {e}")

# 1Ô∏è‚É£ Esperar a que la tabla est√© presente
factura_filas = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//*[@id='invoice-table_wrapper']/div[2]/div/table/tbody/tr")))

# 2Ô∏è‚É£ Listas para almacenar las facturas descargadas y las que fallaron
facturas_descargadas = []
facturas_fallidas = []

# 3Ô∏è‚É£ Iterar sobre cada fila y hacer clic en el bot√≥n de descarga
for index, fila in enumerate(factura_filas, start=1):  
    try:
        # Construir XPath din√°mico para el bot√≥n del men√∫ desplegable (columna 7)
        xpath_boton_menu = f"//*[@id='invoice-table_wrapper']/div[2]/div/table/tbody/tr[{index}]/td[11]/div"

        # Hacer scroll hasta la fila actual
        driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", fila)

        # Esperar a que el bot√≥n del men√∫ sea interactuable y hacer clic
        boton_menu = wait.until(EC.element_to_be_clickable((By.XPATH, xpath_boton_menu)))

        # Usar ActionChains para evitar "element click intercepted"
        ActionChains(driver).move_to_element(boton_menu).click().perform()

        # üîπ Esperar hasta que el men√∫ est√© completamente cargado
        time.sleep(2)  # Peque√±o delay para asegurar que el men√∫ se abre

        # üîπ Esperar a que aparezca la opci√≥n de descarga CSV (usando XPath flexible)
        opcion_csv = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[contains(@id, 'dropdown-download-inv') and contains(@id, '-CSV')]")))

        # Hacer clic en la opci√≥n de descarga CSV
        opcion_csv.click()

        # Peque√±a pausa para evitar sobrecarga de clics
        time.sleep(45)  # 5 segundos en lugar de 30, ajusta seg√∫n sea necesario

        # Capturar el n√∫mero de factura (que est√° dentro de un <button> en la fila)
        numero_factura = fila.find_element(By.XPATH, ".//button[contains(@class, 'btn-datatable-text-link')]").text

        # Guardar la factura descargada
        facturas_descargadas.append(numero_factura)
        print(f"‚úÖ Factura {numero_factura} descargada correctamente.")

    except Exception as e:
        # Si ocurre un error, registrar la factura en la lista de fallidas
        try:
            numero_factura = fila.find_element(By.XPATH, ".//button[contains(@class, 'btn-datatable-text-link')]").text
        except:
            numero_factura = f"Fila {index} (No disponible)"
        
        facturas_fallidas.append(numero_factura)
        print(f"‚ö†Ô∏è Error al descargar la factura {numero_factura}: {e}")

# 4Ô∏è‚É£ Enviar un correo con los resultados al finalizar
send_email(facturas_descargadas, facturas_fallidas)

# 5Ô∏è‚É£ Confirmar que el proceso se complet√≥
print("‚úÖ Descarga completada y correo enviado.")
```
